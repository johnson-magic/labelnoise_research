# 2. learning with label noise
## 2.1 概念
noise label即不可靠的label。有文献统计现实世界中数据集的噪音比例的范围在8.0%~38.5%之间。带噪学习，是指不做显式的数据清洗。而是在带有噪音的情况下进行学习。

## 2.2 现象

由研究表明，深度神经网络可以拟合含有任意噪音比例的整个数据集。如上图左显示，三种不同情形的训练accuracy都可以几乎达到100%。但如果在测试集中出现了与训练accuracy较大的gap，则说明过拟合了。过拟合是指模型学习到了训练集中细节信息和noise。这些信息不具有通用性，因此在面对新的数据时，性能会下降。
